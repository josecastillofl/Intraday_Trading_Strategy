### Download S&P500 Data

```python
# Since there is multiple tables, we only need the first one
sp500 = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]
```

```python
sp500['Symbol'] = sp500['Symbol'].str.replace('.','-')
```

```python
stocks_list = sp500['Symbol'].unique().tolist()
```

```python
end_date = '2023-11-21'
```

```python
start_date = pd.to_datetime(end_date)-pd.DateOffset(365*8)
```

```python
data = yf.download(tickers = stocks_list,
                 start = start_date,
                 end = end_date)
```
